# Classification Model based on Iris Dataset Based on KNC

```python
from sklearn.datasets import load_iris 

pure_df, target = load_iris(as_frame=True, return_X_y=True) 

df = pure_df.copy()
df['target'] = target 

display(df) 
```

### Using PCA to analyse Iris Groups Quantity

```python
from sklearn.decomposition import PCA 
 
df_resultado_fit_transform = PCA(n_components=3).fit_transform(pure_df) 

df_resultado_fit_transform
```

```python
from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler 
import matplotlib.pyplot as plt 
import numpy as np 

standarized_df = StandardScaler().fit_transform(pure_df) 
pca = PCA().fit(standarized_df) 

plt.figure(figsize=(15, 5))
plt.plot(pca.explained_variance_ratio_, color='r')
plt.xticks(np.arange(df.shape[1]))
plt.show() 
```

Percebemos que podemos ter 3 grupos com os dados usando Elbow method para anÃ¡lise

```python
pca_result = PCA(n_components=3).fit_transform(pure_df)
```

```python
from sklearn.feature_selection import SelectKBest 


kbest = SelectKBest(k=4) 
kbest.fit_transform(pure_df, target)
```

```python
from sklearn.model_selection import train_test_split

X = pure_df.values[:,0:4]
Y = target

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score

print(f'Total Accuracy is: {str(accuracy_score(y_test, y_pred) * 100)} %')
```

```python
predictions = model.predict(
    [
        [5.1, 3.5, 1.4, 0.2],
        [5.9, 3.0, 5.1, 1.8],
        [6.9, 3.1, 4.9, 1.5]
    ]
)

print(predictions)
```
